```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .csv sin indice
data= pd.read_csv('gastos_2023_sin_nulos.csv') 
```

```python
#Verificamos informaci√≥n del DataFrame
data.info()
```

```python
#Corroboramos valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

**PROCEDIMIENTO PARA ELIMINAR VALORES NULOS EN DATAFRAME**

```python
#Reemplazamos valores nulos del dataframe con "bfill"
data1 =data.fillna(method="bfill") 
data1
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores nulos del dataframe con "mean"
#Realizamos una copia del dataframe
data2=data.copy()
data2=data2.fillna(round(data.mean(),1))
data2
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
data2.plot(kind='box', vert=False)
plt.title("Valores At√≠picos del Dataframe")
plt.show() #dibujamos el diagrama
```

**PROCEDIMIENTO "DESVIACI√ìN EST√ÅNDAR" PARA ELIMINAR OUTLIERS EN DATAFRAME**

```python
#M√©todo aplicando desviaci√≥n estandar. Encuentro los valores extremos
y=data2
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers del Dataframe
outliers= data2[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
data3= data2[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data3
```

```python
#Eliminamos columnas innecesarias
data4= data3.drop(["FECHA","FOLIO","UUID","TC","RFC","TIPO GASTO","DESCRIPCION","TIPO","STATUS","POLIZA"], axis=1)
data4
```

```python
#Corroboramos valores nulos del dataframe4
valores_nulos=data4.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores at√≠picos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=data4.copy()
data_clean=data_clean.fillna(round(data4.mean(),1))
data_clean
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
# Unimos las columnas con el dataframe

cols=data2["FECHA"]
cols1=data2["FOLIO"]
cols2=data2["UUID"]
cols3=data2["TC"]
cols5=data2["RFC"]
cols6=data2["TIPO GASTO"]
cols8=data2["DESCRIPCION"]
cols9=data2["TIPO"]
cols10=data2["STATUS"]
cols11=data2["POLIZA"]

#"FOLIO","UUID","TC","Columna1","GASTO","RFC","PROVEEDOR","TIPO GASTO","DESCRIPCION","TIPO","STATUS","POLIZA"
Datos_limpios = pd.concat([cols,cols1,cols2,cols3,cols5,cols6,cols8,cols9,cols10,cols11,data_clean], axis=1)
Datos_limpios
```

```python
#Convertir DataFrame a CSV
Datos_limpios.to_csv("Datos_limpios2023_des.csv")

```

**PROCEDIMIENTO "CUANTILES" PARA SUSTITUIR OUTLIERS EN DATAFRAME**

```python
#M√©todo aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=data2

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers del Dataframe
outliers_iqr= data2[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios del Dataframe
data3_iqr= data2[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
#Eliminamos columnas innecesarias
data4_iqr= data3_iqr.drop(["FECHA","FOLIO","UUID","TC","RFC","TIPO GASTO","DESCRIPCION","TIPO","STATUS","POLIZA"], axis=1)
data4_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data4_iqr.isnull().sum()
valores_nulos
```
